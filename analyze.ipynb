```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Tc.csv')
result = df.iloc[:, 0:2]
print(result)
```

```python
plt.plot(result["Size"], result["Tc_estimated"], color = "#FF0000", marker="o")
T_theory = 2.269
plt.axhline(T_theory, linestyle="--", color = "#FF00FF", label="Theoretical Tc≈2.269")

plt.xlabel("Lattice size")
plt.ylabel("Estimated critical temperature")
plt.legend()
plt.show()
```
